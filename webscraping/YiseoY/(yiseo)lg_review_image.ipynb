```python
import os

folder_path = '/content/drive/Othercomputers/My Mac/Workspaces/SKN14-3rd-3Team/webscraping/YiseoY/lg_img'
unique_names = set()

for root, dirs, files in os.walk(folder_path):
    for file in files:
        # 확장자 제거
        name, _ = os.path.splitext(file)

        # 맨 마지막이 _숫자일 경우 제거
        if '_' in name:
            parts = name.split('_')
            if parts[-1].isdigit():
                name = '_'.join(parts[:-1])

        unique_names.add(name)  # set에 추가 (중복 자동 제거)

# 중복 제거된 이름들을 정렬하여 리스트로 변환
result_names = sorted(unique_names)

# 결과 출력
for name in result_names:
    print(name)

print(f'총 개수: {len(result_names)}')
```

```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager
import time

# 1. 크롤러(Driver) 관리 클래스
class WebDriverManager:
    def __init__(self):
        options = Options()
        options.add_argument('--headless=new')
        options.add_argument('--no-sandbox')
        options.add_argument('--disable-dev-shm-usage')
        self.driver = webdriver.Chrome(options=options)

    def quit(self):
        self.driver.quit()

class ProductReviewPage:
    def __init__(self, driver, url):
        self.driver = driver
        self.url = url
        self.wait = WebDriverWait(self.driver, 20)

    def open(self):
        self.driver.get(self.url)
        time.sleep(1)

    def click_tab(self):
        try:
            # 'data-link-name'이 'review'인 a 태그 직접 찾기
            review_a = self.driver.find_element(By.CSS_SELECTOR, "div.contents-menu-list a[data-link-name='review']")
            self.driver.execute_script("arguments[0].scrollIntoView(true);", review_a)
            self.driver.execute_script("arguments[0].click();", review_a)
            time.sleep(1)
            return True
        except Exception as e:
            print(f"    ❌ 리뷰 탭 클릭 실패: {e}")
            return False

    def get_review_img(self):
        try:
            if self.click_tab():
                # 리뷰 영역 로딩 대기
                self.wait.until(EC.presence_of_element_located((By.CLASS_NAME, "product-review-wrap")))
                img_srcs = []
                # 1. 리뷰 wrap들 찾기
                wraps = self.driver.find_elements(By.CSS_SELECTOR, "div.product-review-wrap")
                for wrap in wraps:
                    # 2. 리뷰 리스트(ul) 찾기
                    review_uls = wrap.find_elements(By.CSS_SELECTOR, "ul.product-review-list")
                    for review_ul in review_uls:
                        # 3. 각 리뷰 li 찾기
                        review_lis = review_ul.find_elements(By.TAG_NAME, "li")
                        for review_li in review_lis:
                            # 4. media-list ul 내부 li 반복
                            media_uls = review_li.find_elements(By.CSS_SELECTOR, "ul.media-list")
                            for media_ul in media_uls:
                                media_lis = media_ul.find_elements(By.TAG_NAME, "li")
                                for media_li in media_lis:
                                    # 5. thumb div 내부 img 찾기
                                    thumb_divs = media_li.find_elements(By.CSS_SELECTOR, "div.thumb")
                                    for thumb in thumb_divs:
                                        img = thumb.find_element(By.TAG_NAME, "img")
                                        src = img.get_attribute('src')
                                        if src:
                                            img_srcs.append(src)
                return img_srcs
            else:
                return []
        except Exception as e:
            print(f"    ❌ 리뷰 이미지 수집 실패: {e}")
            return []
```

```python
# WebDriver 설정
manager = WebDriverManager()
driver = manager.driver

# 결과를 여기에 저장
review_img_column = []

for url in unique_df['product_url']:
    try:
        print(f"🔍 크롤링 중: {url}")
        page = ProductReviewPage(driver, url)
        page.open()
        img_srcs = page.get_review_img()
        review_img_column.append(img_srcs)
        print(f"   👉 수집 이미지 수: {len(img_srcs)}")
    except Exception as e:
        print(f"   ❌ 에러 발생: {e}")
        review_img_column.append([])

# 최종 결과를 df에 추가
unique_df['review_img'] = review_img_column

# 크롬 드라이버 종료
manager.quit()
```

```python
import os
import requests
import re
from tqdm import tqdm

# ✅ 저장 디렉토리 설정
save_dir = "/content/drive/Othercomputers/My Mac/Workspaces/SKN14-3rd-3Team/webscraping/YiseoY/lg_reviews"
os.makedirs(save_dir, exist_ok=True)

# 🔧 파일명에서 위험한 문자 제거하는 함수
def sanitize_filename(text):
    return re.sub(r'[\\/*?:"<>|]', '_', text)

# ✅ 모델별 카운터 딕셔너리 (model_name: next_index)
model_counters = {}

# ✅ 수집 및 저장 루프
for idx, row in tqdm(unique_df.iterrows(), total=len(unique_df)):
    # 모델명 정제
    model_name = sanitize_filename(str(row['model_name']).strip().replace(" ", "_"))

    # 이미지 리스트
    review_list = row['review_img']
    if not isinstance(review_list, list) or len(review_list) == 0:
        continue

    # 모델별 카운터 초기화
    if model_name not in model_counters:
        model_counters[model_name] = 1

    # 이미지 순차 저장
    for review_url in review_list:
        try:
            response = requests.get(review_url, timeout=10)
            if response.status_code == 200:
                # 카운터 가져오기
                count = model_counters[model_name]

                # 파일명 구성: model_name_0001.png
                file_name = f"{model_name}_{count:04d}.png"
                file_path = os.path.join(save_dir, file_name)

                # 이미지 저장
                with open(file_path, "wb") as f:
                    f.write(response.content)

                print(f"✅ 저장됨: {file_name}")

                # 카운터 증가
                model_counters[model_name] += 1
            else:
                print(f"⚠️ 응답 오류({response.status_code}): {review_url}")
        except Exception as e:
            print(f"❌ 다운로드 실패: {review_url} | 오류: {e}")
```
