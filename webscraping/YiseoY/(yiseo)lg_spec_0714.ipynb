```python
import time
import json
import pandas as pd
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options

# 1. 스크롤 함수
def scroll_through_all_lis(driver):
    last_height = driver.execute_script("return document.body.scrollHeight")
    while True:
        lis = driver.find_elements(By.CSS_SELECTOR, "section.CommonPcListUnitProduct_list_unit_product__AeuOT ul[role='list'] > li")
        for li in lis:
            try:
                driver.execute_script("arguments[0].scrollIntoView(true);", li)
                time.sleep(0.3)
            except:
                continue
        new_height = driver.execute_script("return document.body.scrollHeight")
        if new_height == last_height:
            break
        last_height = new_height

# 2. 베이스 URL 추출
def get_base_url(url):
    from urllib.parse import urlparse
    parsed = urlparse(url)
    return f"{parsed.scheme}://{parsed.netloc}"

# 3. 상품 리스트 + 상세페이지 링크 추출
def scrape_products_with_links(driver, url):
    driver.get(url)
    time.sleep(3)
    scroll_through_all_lis(driver)
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    section = soup.find('section', class_="CommonPcListUnitProduct_list_unit_product__AeuOT")
    products = []
    base_url = get_base_url(url)
    if section:
        ul = section.find('ul', role='list')
        if ul:
            lis = ul.find_all('li')
            for li in lis:
                data = li.get('data-ec-product')
                a_tag = li.find('a', href=True)
                link = a_tag['href'] if a_tag else None
                if data and link:
                    products.append({
                        'ec_product': data,
                        'detail_link': base_url + link if link.startswith('/') else base_url + '/' + link
                    })
    return products

# 4. 상세페이지 사양 크롤링
def scrape_lg_washer_specs(detail_url):
    options = Options()
    options.add_argument('--headless')
    options.add_argument('--no-sandbox')
    options.add_argument('--disable-dev-shm-usage')
    driver = webdriver.Chrome(options=options)
    driver.get(detail_url)
    time.sleep(2)
    try:
        btn = driver.find_element(By.CLASS_NAME, "btn_collapse_more")
        a_tag = btn.find_element(By.TAG_NAME, "a")
        a_tag.click()
        time.sleep(2)
        spec_detail = driver.find_element(By.CLASS_NAME, "prod-spec-detail")
        box_elements = spec_detail.find_elements(By.CLASS_NAME, 'box')
        specs = []
        for box in box_elements:
            try:
                h3 = box.find_element(By.TAG_NAME, 'h3')
                h3_text = h3.text.strip()
            except Exception:
                h3_text = None
            try:
                spec_list = box.find_element(By.CLASS_NAME, 'spec-info-list')
                ul_element = spec_list.find_element(By.TAG_NAME, 'ul')
                li_elements = ul_element.find_elements(By.TAG_NAME, 'li')
            except Exception:
                li_elements = []
            for li in li_elements:
                dl = li.find_element(By.TAG_NAME, 'dl')
                dt_elem = dl.find_element(By.TAG_NAME, 'dt')
                dd_elem = dl.find_element(By.TAG_NAME, 'dd')
                dt_text = dt_elem.text.strip()
                dd_text = dd_elem.text.strip()
                desc = None
                try:
                    button = dt_elem.find_element(By.TAG_NAME, 'button')
                    desc = button.get_attribute('data-spec-description')
                except Exception:
                    pass
                specs.append({
                    'box_title': h3_text,
                    'spec_name': dt_text,
                    'spec_value': dd_text,
                    'spec_description': desc
                })
        driver.quit()
        return specs
    except Exception as e:
        print(f"오류 발생: {e}")
        driver.quit()
        return None

# 5. 전체 카테고리 순회 및 상세페이지 크롤링
urls = [
    "https://www.lge.co.kr/category/wash-tower",
    "https://www.lge.co.kr/category/wash-combo",
    "https://www.lge.co.kr/category/washing-machines",
    "https://www.lge.co.kr/category/dryers"
]

options = Options()
options.add_argument('--headless')
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')
main_driver = webdriver.Chrome(options=options)

all_data = []

for url in urls:
    print(f"\n크롤링 중: {url}")
    products = scrape_products_with_links(main_driver, url)
    for prod in products:
        try:
            item = json.loads(prod['ec_product'])
            sku = item.get('model_sku', '').replace('.AKOR', '').strip()
            detail_url = prod['detail_link']
            print(f"\n- 상세페이지 크롤링: {detail_url} (SKU: {sku})")
            specs = scrape_lg_washer_specs(detail_url)
            all_data.extend(specs)
            if specs:
                for spec in specs:
                    print(spec)
            else:
                print("사양 정보를 가져오지 못했습니다.")
        except Exception as e:
            print(f"상품 정보 파싱 실패: {prod} ({e})")

# 6. 결과 CSV로 저장
df = pd.DataFrame(all_data)
df.to_csv("크롤링_결과.csv", index=False, encoding='utf-8-sig')
print("✅ 크롤링 결과가 '크롤링_결과.csv'로 저장되었습니다.")
```
