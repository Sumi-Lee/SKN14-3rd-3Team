```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager
import time

# 1. 크롤러(Driver) 관리 클래스
class WebDriverManager:
    def __init__(self):
        options = Options()
        options.add_argument('--headless=new')
        options.add_argument('--no-sandbox')
        options.add_argument('--disable-dev-shm-usage')
        self.driver = webdriver.Chrome(options=options)

    def quit(self):
        self.driver.quit()

# 2. 제품 상세 페이지 객체 클래스
class ProductPage:
    def __init__(self, driver, url):
        self.driver = driver
        self.url = url
        self.wait = WebDriverWait(self.driver, 10)

    def open(self):
        self.driver.get(self.url)
        time.sleep(1)  # 페이지 로딩 대기

    def open_image_popup(self):
        # '추가 이미지' 버튼 클릭
        more_btn = self.wait.until(EC.element_to_be_clickable((By.CSS_SELECTOR, "li.thumbnail.more > a")))
        self.driver.execute_script("arguments[0].scrollIntoView(true);", more_btn)
        time.sleep(1)
        self.driver.execute_script("arguments[0].click();", more_btn)
        # 팝업창 대기
        self.wait.until(EC.visibility_of_element_located((By.CLASS_NAME, "ui_modal_wrap")))

    def get_popup_image_urls(self):
        # 팝업 내 이미지 추출
        modal = self.driver.find_element(By.CLASS_NAME, "ui_modal_wrap")
        imgs = modal.find_elements(By.CSS_SELECTOR, "img")
        return [img.get_attribute("src") for img in imgs]

# 3. 데이터프레임 순회 및 결과 저장
def collect_all_images(product_urls):
    manager = WebDriverManager()
    driver = manager.driver
    results = []
    for url in product_urls:
        print(f"Processing: {url}")
        try:
            page = ProductPage(driver, url)
            page.open()
            page.open_image_popup()
            img_urls = page.get_popup_image_urls()
            print(f"  Found {len(img_urls)} images.")
        except Exception as e:
            print(f"  Failed: {e}")
            img_urls = []
        results.append(img_urls)
    manager.quit()
    return results


filtered_df['image_url'] = collect_all_images(filtered_df['product_url'])
print(filtered_df[['product_url', 'image_url']].head())
```

```python
import os
import requests
from tqdm import tqdm

save_dir = "/content/drive/Othercomputers/My Mac/Workspaces/SKN14-3rd-3Team/webscraping/YiseoY/lg_img_classifier_output"
os.makedirs(save_dir, exist_ok=True)

prev_model_sku = None
image_counter = 1

for idx, row in tqdm(filtered_df.iterrows(), total=len(filtered_df)):
    model_name = str(row['model_name']).strip().replace(" ", "_")
    model_sku = str(row['model_sku']).strip()
    img_urls = row['image_url']

    # model_sku가 바뀌면 넘버링 초기화
    if model_sku != prev_model_sku:
        image_counter = 1
        prev_model_sku = model_sku

    if not isinstance(img_urls, list):
        continue

    for url in img_urls:
        try:
            directory_name = model_name + "_" + model_sku
            directory_dir = os.path.join(save_dir, directory_name)
            os.makedirs(directory_dir, exist_ok=True)
            file_name = f"{model_name}_{model_sku}_{image_counter:04d}.jpg"
            file_path = os.path.join(directory_dir, file_name)

            response = requests.get(url, timeout=10)
            if response.status_code == 200:
                with open(file_path, "wb") as f:
                    f.write(response.content)
                image_counter += 1
            else:
                print(f"⚠️ 다운로드 실패 (status {response.status_code}): {url}")

        except Exception as e:
            print(f"❌ 오류 발생: {url} | {e}")
```
