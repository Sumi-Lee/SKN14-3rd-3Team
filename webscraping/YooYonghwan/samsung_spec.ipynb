```python
# 더보기버튼 수행 (품절상품 제외 해제기능 없음)

import os
import time
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# ChromeDriver 경로 설정 (노트북 파일과 동일한 디렉토리에 있다고 가정)
chrome_driver_path = 'chromedriver.exe'

# 크롬 옵션 설정
chrome_options = Options()
chrome_options.add_argument('--disable-extensions')
# 헤드리스 모드 비활성화
# chrome_options.add_argument('--headless')

# 팝업 허용 설정
prefs = {
    'profile.default_content_setting_values.notifications': 1,
    'profile.default_content_settings.popups': 0,
    'profile.default_content_setting_values.automatic_downloads': 1
}
chrome_options.add_experimental_option('prefs', prefs)

# 크롬 드라이버 서비스 설정
service = Service(executable_path=chrome_driver_path)

# 크롬 드라이버 실행
driver = webdriver.Chrome(service=service, options=chrome_options)

# 크롤링 대상 URL
url = "https://www.samsung.com/sec/washers-and-dryers/all-washers-and-dryers/"

# 페이지 로딩 대기 시간 (초)
WAIT_TIME = 15

# 1. 해당 URL로 접속
try:
    driver.get(url)
    print(f"페이지 접속 완료: {url}")
except Exception as e:
    print(f"페이지 접속 실패: {e}")
    driver.quit()
    exit()

# 페이지 로딩이 완료될 때까지 대기
try:
    WebDriverWait(driver, WAIT_TIME).until(lambda driver: driver.execute_script('return document.readyState') == 'complete')
    print("페이지 로딩 완료")
except Exception as e:
    print(f"페이지 로딩 실패: {e}")
    driver.quit()
    exit()

# 2. 더보기 버튼 확인 및 감지
while True:
    try:
        more_button = WebDriverWait(driver, WAIT_TIME).until(EC.presence_of_element_located((By.CSS_SELECTOR, "button.btn.btn-type1.btn-d.btn-readmore")))
        print("더보기 버튼 감지")

        # 현재 페이지 번호와 전체 페이지 번호 확인
        current_page = int(driver.find_element(By.ID, "presentPageCount").text)
        total_pages = int(driver.find_element(By.ID, "totalPageCount").text)
        print(f"현재 페이지: {current_page}/{total_pages}")

        if current_page == total_pages:
            print("마지막 페이지입니다.")
            break

        # 3. 클릭
        more_button.click()
        print("더보기 버튼 클릭")
        time.sleep(2)  # 추가 로딩 대기

        # 4. 스크롤
        driver.execute_script("arguments[0].scrollIntoView();", more_button)
        print("스크롤 이동")

    except Exception as e:
        print(f"더보기 버튼 찾기 실패: {e}")
        break

# 6. 스크롤이 맨 아래쪽으로 내려가면 종료
while True:
    # 현재 스크롤 위치
    current_position = driver.execute_script("return window.pageYOffset;")
    # 전체 페이지 높이
    page_height = driver.execute_script("return document.body.scrollHeight;")

    # 스크롤을 맨 아래로 내림
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(2)  # 스크롤 이동 후 대기

    # 스크롤 위치 확인
    new_position = driver.execute_script("return window.pageYOffset;")

    if new_position == current_position:
        print("스크롤이 맨 아래쪽에 도달했습니다.")
        break

# 크롬 드라이버 종료
driver.quit()
```

```python
# 품절상품 제외 해제 + 더보기 버튼 클리 -> 모든 제품이 다 나오도록 함 (총 102개)
import os
import time
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# ChromeDriver 경로 설정 (노트북 파일과 동일한 디렉토리에 있다고 가정)
chrome_driver_path = 'chromedriver.exe'

# 크롬 옵션 설정
chrome_options = Options()
chrome_options.add_argument('--disable-extensions')
# 헤드리스 모드 비활성화
# chrome_options.add_argument('--headless')

# 팝업 허용 설정
prefs = {
    'profile.default_content_setting_values.notifications': 1,
    'profile.default_content_settings.popups': 0,
    'profile.default_content_setting_values.automatic_downloads': 1
}
chrome_options.add_experimental_option('prefs', prefs)

# 크롬 드라이버 서비스 설정
service = Service(executable_path=chrome_driver_path)

# 크롬 드라이버 실행
driver = webdriver.Chrome(service=service, options=chrome_options)

# 크롤링 대상 URL
url = "https://www.samsung.com/sec/washers-and-dryers/all-washers-and-dryers/"

# 페이지 로딩 대기 시간 (초)
WAIT_TIME = 15

# 1. 해당 URL로 접속
try:
    driver.get(url)
    print(f"페이지 접속 완료: {url}")
except Exception as e:
    print(f"페이지 접속 실패: {e}")
    driver.quit()
    exit()

# 페이지 로딩이 완료될 때까지 대기
try:
    WebDriverWait(driver, WAIT_TIME).until(lambda driver: driver.execute_script('return document.readyState') == 'complete')
    print("페이지 로딩 완료")
except Exception as e:
    print(f"페이지 로딩 실패: {e}")
    driver.quit()
    exit()

# 품절상품 제외 체크박스 감지 및 클릭
try:
    exclude_sold_out_label = WebDriverWait(driver, WAIT_TIME).until(EC.presence_of_element_located((By.CSS_SELECTOR, "label[for='toggle-check-on-pc']")))
    print("품절상품 제외 체크박스 라벨 감지")

    exclude_sold_out_checkbox = driver.find_element(By.CSS_SELECTOR, "input#toggle-check-on-pc")

    if exclude_sold_out_checkbox.is_selected():
        exclude_sold_out_label.click()
        print("품절상품 제외 체크박스 해제")
    else:
        exclude_sold_out_label.click()
        print("품절상품 제외 체크박스 선택")

    time.sleep(2)  # 클릭 후 대기
except Exception as e:
    print(f"품절상품 제외 체크박스 라벨 찾기 실패: {e}")


# 2. 더보기 버튼 확인 및 감지
while True:
    try:
        more_button = WebDriverWait(driver, WAIT_TIME).until(EC.presence_of_element_located((By.CSS_SELECTOR, "button.btn.btn-type1.btn-d.btn-readmore")))
        print("더보기 버튼 감지")

        # 현재 페이지 번호와 전체 페이지 번호 확인
        current_page = int(driver.find_element(By.ID, "presentPageCount").text)
        total_pages = int(driver.find_element(By.ID, "totalPageCount").text)
        print(f"현재 페이지: {current_page}/{total_pages}")

        if current_page == total_pages:
            print("마지막 페이지입니다.")
            break

        # 3. 자바스크립트를 사용하여 더보기 버튼 클릭
        driver.execute_script("arguments[0].click();", more_button)
        print("더보기 버튼 클릭")
        time.sleep(2)  # 추가 로딩 대기

        # 4. 스크롤
        driver.execute_script("arguments[0].scrollIntoView();", more_button)
        print("스크롤 이동")

    except Exception as e:
        print(f"더보기 버튼 찾기 실패: {e}")
        break

# 6. 스크롤이 맨 아래쪽으로 내려가면 종료
while True:
    # 현재 스크롤 위치
    current_position = driver.execute_script("return window.pageYOffset;")
    # 전체 페이지 높이
    page_height = driver.execute_script("return document.body.scrollHeight;")

    # 스크롤을 맨 아래로 내림
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(2)  # 스크롤 이동 후 대기

    # 스크롤 위치 확인
    new_position = driver.execute_script("return window.pageYOffset;")

    if new_position == current_position:
        print("스크롤이 맨 아래쪽에 도달했습니다.")
        break

# 제품 URL 링크 수집
from bs4 import BeautifulSoup
from urllib.parse import urljoin

base_url = "https://www.samsung.com"

# 현재 페이지의 HTML 소스 가져오기
html = driver.page_source

# BeautifulSoup을 사용하여 HTML 파싱
soup = BeautifulSoup(html, 'html.parser')

# 상품 URL 수집
product_links = []
product_elements = soup.select('div.item-inner > div.card-detail > a')
for product_element in product_elements:
    relative_link = product_element['href']
    absolute_link = urljoin(base_url, relative_link.split("'")[1])
    product_links.append(absolute_link)
    print(f"제품 링크: {absolute_link}")

print(f"총 {len(product_links)}개의 제품 링크를 수집했습니다.")


import json
import csv
from bs4 import BeautifulSoup
import time

# ✅ 1. 제품 상세 정보 추출 함수 — 예시 URL 구조 기반 (h2, span, price-em, price-dc, img)
def scrape_product_detail(url):
    driver.get(url)
    time.sleep(2)

    # ✅ 디버깅용 HTML 저장
    with open("debug_page.html", "w", encoding="utf-8") as f:
        f.write(driver.page_source)

    soup = BeautifulSoup(driver.page_source, 'html.parser')

    model_name_ko = soup.select_one("div.product-intro__title h1")  # 한글 이름
    model_name_ko = model_name_ko.get_text(strip=True) if model_name_ko else None

    model_name_en = soup.select_one("div.product-intro__subtitle")  # 영어 모델명
    model_name_en = model_name_en.get_text(strip=True) if model_name_en else None

    price_elem = soup.select_one("div.product-intro__price em")
    price = price_elem.get_text(strip=True).replace(",", "") if price_elem else None

    discount_elem = soup.select_one("div.product-intro__price span.sale-price em")
    discounted_price = discount_elem.get_text(strip=True).replace(",", "") if discount_elem else price

    img_elem = soup.select_one("div.product-image-gallery img")
    img_url = img_elem["src"] if img_elem else None

    print("----- 제품 정보 추출 -----")
    print(f"한글 모델명 : {model_name_ko}")
    print(f"영어 모델명 : {model_name_en}")
    print(f"가격       : {price}")
    print(f"할인가   : {discounted_price}")
    print(f"IMG URL    : {img_url}")
    print("-------------------------")

    return {
        "model_name_ko": model_name_ko,
        "model_name_en": model_name_en,
        "price": price,
        "discounted_price": discounted_price,
        "img_url": img_url,
        "url": url
    }

# ✅ 2. 모든 product_links 순회하여 정보 수집 + PRINT + 저장 리스트 구축
all_data = []
for idx, link in enumerate(product_links, 1):
    print(f"\n[{idx}/{len(product_links)}] 크롤링 중: {link}")
    try:
        info = scrape_product_detail(link)
        all_data.append(info)
    except Exception as e:
        print(f"❌ 오류 발생: {e}")

print(f"\n✅ 총 {len(all_data)}개 제품 데이터 수집 완료")

# ✅ 3. JSON 파일 저장
with open("products.json", "w", encoding="utf-8") as f:
    json.dump(all_data, f, ensure_ascii=False, indent=2)
print("✅ JSON 파일 저장됨: products.json")

# ✅ 4. CSV 파일 저장
if all_data:
    with open("products.csv", "w", encoding="utf-8-sig", newline="") as f:
        writer = csv.DictWriter(f, fieldnames=list(all_data[0].keys()))
        writer.writeheader()
        writer.writerows(all_data)
    print("✅ CSV 파일 저장됨: products.csv")

# ✅ 5. 드라이버 종료
driver.quit()
```
